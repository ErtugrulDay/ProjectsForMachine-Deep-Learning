```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
df=pd.read_csv("polynomial.csv",sep=";")
df
```

```python
plt.scatter(df.deneyim,df.maas)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```

As can be seen, the data is not linearly distribute, let's see this by applying linear regression.

```python
reg=LinearRegression()
reg.fit(df[["deneyim"]],df["maas"])
plt.xlabel("Experience(Year)")
plt.ylabel("Salary")
plt.scatter(df.deneyim,df.maas)
x=df["deneyim"]
y=reg.predict(df[["deneyim"]])
plt.plot(x,y,color="green",label="Linear-Regression")
plt.legend()
plt.show()
```

What if we apply polynomial regression ?

```python
polynomial_regression=PolynomialFeatures(degree=3)
polynomial_regression2=PolynomialFeatures(degree=4)
x_polynomail=polynomial_regression.fit_transform(df[["deneyim"]])
x_polynomail2=polynomial_regression2.fit_transform(df[["deneyim"]])
```

```python
reg=LinearRegression()
reg2=LinearRegression()
reg.fit(x_polynomail,df.maas)
reg2.fit(x_polynomail2,df.maas)
```

```python
y_pred=reg.predict(x_polynomail)
y_pred2=reg2.predict(x_polynomail2)
```

```python
#for degree 3
plt.plot(df.deneyim,y_pred,color="red",label="polynomial-regression, degree=3")
plt.xlabel("Experience(Year)")
plt.ylabel("Salary")
plt.legend()
plt.scatter(df.deneyim,df.maas)
plt.show()
```

```python
#for degree 4
plt.plot(df.deneyim,y_pred2,color="red",label="polynomial-regression, degree=4")
plt.xlabel("Experience(Year)")
plt.ylabel("Salary")
plt.legend()
plt.scatter(df.deneyim,df.maas)
plt.show()
```
